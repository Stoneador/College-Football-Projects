```python
def get_Data(Team):

    team = Team.replace(' ','-')
    team = team.replace('(','').lower()
    team = team.replace(')','')
    team = team.replace('&','')
    team = team.replace('uconn','connecticut')
    team = team.replace('smu','southern-methodist')
    team = team.replace('ucf','central-florida')
    team = team.replace('usf','south-florida')
    team = team.replace('tcu','texas-christian')
    team = team.replace('fiu','florida-international')
    team = team.replace('fau','florida_atlantic')
    team = team.replace('middle-tennessee','middle-tennessee-state')
    team = team.replace('uab','alabama-birmingham')
    team = team.replace('southern-miss','southern-mississippi')
    team = team.replace('utep','texas-el-paso')
    team = team.replace('utsa','texas-san-antonio')
    team = team.replace('byu','brigham-young')
    team = team.replace('umass','massachusetts')
    team = team.replace('bowling-green','bowling-green-state')
    team = team.replace('unlv','nevada-las-vegas')
    team = team.replace('usc','southern-california')
    team = team.replace('lsu','louisiana-state')
    team = team.replace('ole-miss','mississippi')

    url="https://www.sports-reference.com/cfb/schools/{}/index.html".format(team)
    # use urllib and BeautifulSoup in order to open and copy the html code of the webpage to another variable
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'lxml')
    # use soup.find in order to pull out only the html code related to the data wanted
    record_list = soup.findAll('tr')
    #strip = record_list.text.strip()

    A = []
    B = []
    C = []
    D = []
    E = []
    F = []
    G = []
    H = []
    I = []
    J = []
    K = []
    L = []
    M = []
    N = []
    O = []
    P = []
    Q = []
    R = []

    Name = []

    for row in record_list:
        cells = row.findAll('td')
        if len(cells) == 18:
            A.append(cells[0].find(text=True))
            B.append(cells[1].find(text=True))
            C.append(cells[2].find(text=True))
            D.append(cells[3].find(text=True))
            E.append(cells[4].find(text=True))
            F.append(cells[5].find(text=True))
            G.append(cells[6].find(text=True))
            H.append(cells[7].find(text=True))
            I.append(cells[8].find(text=True))
            J.append(cells[9].find(text=True))
            K.append(cells[10].find(text=True))
            L.append(cells[11].find(text=True))
            M.append(cells[12].find(text=True))
            N.append(cells[13].find(text=True))
            O.append(cells[14].find(text=True))
            P.append(cells[15].find(text=True))
            Q.append(cells[16].find(text=True))
            R.append(cells[17].find(text=True))
    l = len(A)
    Name = [Team] * l
    df_Record = pd.DataFrame(Name,columns = ['Team'])
    df_Record['Year'] = A
    df_Record['Conference'] = B
    df_Record['Wins'] = C
    df_Record['Losses'] = D
    df_Record['Ties'] = E
    df_Record['Win %'] = F
    df_Record['C Wins'] = G
    df_Record['C Losses'] = H
    df_Record['C Ties'] = I
    df_Record['C %'] = J
    df_Record['SRS'] = K
    df_Record['SOS'] = L
    df_Record['AP Pre'] = M
    df_Record['AP High'] = N
    df_Record['AP Post'] = O
    df_Record['Year'] = df_Record['Year'].astype(int)


    Diff = []
    for Pre, Post, Year in zip(df_Record['AP Pre'],df_Record['AP Post'], df_Record['Year']):
        if Year > 1988:
            if Pre == None:
                Pre = 26
            if Post == None:
                Post = 26

        if Year > 1967 & Year <= 1988:
            if Pre == None:
                Pre = 21
            if Post == None:
                Post = 21

        if Year > 1961 & Year <= 1967:
            if Pre == None:
                Pre = 11
            if Post == None:
                Post = 11

        if Year > 1949 & Year <= 1961:
            if Pre == None:
                Pre = 21
            if Post == None:
                Post = 21
        if Year < 1950:
            Diff.append('None')
        else:
            Diff.append(int(int(Pre) - int(Post)))

        Name.append(Team)

    df_Record['AP Diff'] = Diff
    df_Record['Coach(es)'] = P
    df_Record['Bowl'] = Q
    df_Record['Notes'] = R

    #df_Record['AP Diff'] = df_Record['AP Diff'].astype(int)



    return(df_Record)
```

```python
def get_List(Team, df_Record):

    List = [Team]
    O_5 = 0
    U_5 = 0
    E_5 = 0
    O_20 = 0
    U_20 = 0
    E_20 = 0
    O = 0
    U = 0
    E = 0

    Last_5 = df_Record[(df_Record['Year'] > 2015) & (df_Record['Year'] < 2021)]
    Mean_5 = mean(Last_5['AP Diff'].tolist())
    List.append(Mean_5)

    for diff in Last_5['AP Diff']:
        if diff == None:
            E_5 = E_5 + 1
        elif int(diff) < 0:
            O_5 = O_5 + 1
        elif int(diff) > 0:
            U_5 = U_5 + 1
        elif int(diff) == 0:
            E_5 = E_5 + 1
    Stat_5 = str(O_5)+"/"+str(U_5)+"/"+str(E_5)
    List.append(Stat_5)

    Last_20 = df_Record[(df_Record['Year'] > 2000) & (df_Record['Year'] < 2021)]
    Mean_20 = mean(Last_20['AP Diff'].tolist())
    List.append(Mean_20)

    for diff in Last_20['AP Diff']:
        if diff == None:
            E_20 = E_20 + 1
        elif int(diff) < 0:
            O_20 = O_20 + 1
        elif int(diff) > 0:
            U_20 = U_20 + 1
        elif int(diff) == 0:
            E_20 = E_20 + 1
    Stat_20 = str(O_20)+"/"+str(U_20)+"/"+str(E_20)
    List.append(Stat_20)

    All_Time = df_Record[(df_Record['Year'] > 1949) & (df_Record['Year'] < 2021)]
    Mean = mean(All_Time['AP Diff'].tolist())
    List.append(Mean)

    for diff in All_Time['AP Diff']:
        if diff == None:
            E = E + 1
        elif int(diff) < 0:
            O = O + 1
        elif int(diff) > 0:
            U = U + 1
        elif int(diff) == 0:
            E = E + 1
    Stat = str(O)+"/"+str(U)+"/"+str(E)
    List.append(Stat)

    return(List)
```
